```python
import pandas as pd
```

```python
data = pd.read_csv('data/raw/apartments_for_rent_classified_100K.csv', delimiter=';',encoding="ISO-8859-1")
data.head()
```

```python
data = data.sort_values(by=['time'])
data.head()
```

```python
def split_data(data):
    df_1 = data[:int(len(data)/5)]
    df_2 = data[int(len(data)/5):int(len(data)/5)*2]
    df_3 = data[int(len(data)/5)*2:int(len(data)/5)*3]
    df_4 = data[int(len(data)/5)*3:int(len(data)/5)*4]
    df_5 = data[int(len(data)/5)*4:]

    return df_1, df_2, df_3, df_4, df_5
```

```python
df_1, df_2, df_3, df_4, df_5 = split_data(data)
```

```python
df_1.shape[0] + df_2.shape[0] + df_3.shape[0] + df_4.shape[0] + df_5.shape[0] == data.shape[0]
```

```python
list_df = [df_1, df_2, df_3, df_4, df_5]
for i, df in enumerate(list_df):
    df.to_csv('data/raw/split_data/df_'+str(i+1)+'.csv', index=False)
    
```
